```python
weather = pd.read_csv('./data/weather.csv')
from pandas import DataFrame as df

weather['일시'] = pd.to_datetime(weather['일시'], format='%Y%m%d%H')
```

```python
df1 = df(data={'year':[], 'month':[], 'day':[], 'hour':[]})

df1['year'] = weather['일시'].astype(str).str[0:4]
df1['month'] = weather['일시'].astype(str).str[5:7]
#month = month.rename(columns={'일시':'month'})
df1['day'] = weather['일시'].astype(str).str[8:10]
df1['hour'] = weather['일시'].astype(str).str[11:13]

df1
```

```python
weather = pd.concat([weather, df1], axis=1)

def season(month):
    if month>=3 and month<=5: return 0
    elif month>=6 and month<=8: return 1
    elif month>=9 and month<=11: return 2
    else: return 3

# print(weather.head(5))
# weather['season'] = 0
# print(weather.iloc[0, 8])
ss = []
for i in weather['month']:
    ss.append(season(int(i)))
    
```

```python
weather['season'] = ss

weather
```
